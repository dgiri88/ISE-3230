```python
from itertools import permutations

def are_cyclic_permutations(arr1, arr2):
    if len(arr1) != len(arr2):
        return False

    n = len(arr1)

    # Check if arr2 is a cyclic permutation of arr1
    for i in range(n):
        if arr1 == arr2[i:] + arr2[:i]:
            return True

    return False

def generate_non_cyclic_permutations(length, x):
    if x < 0 or x > length:
        raise ValueError("Invalid value for x")

    # Generate all permutations of a binary array of length with x ones
    binary_array = [1] * x + [0] * (length - x)
    all_permutations = list(permutations(binary_array))

    # Initialize a set with the first permutation
    non_cyclic_permutations = set([all_permutations[0]])

    # Iterate over the remaining permutations
    for i in range(1, len(all_permutations)):
        array2 = all_permutations[i]

        # Check if array2 is not a cyclic permutation of any array in the set
        if all(not are_cyclic_permutations(array1, array2) for array1 in non_cyclic_permutations):
            non_cyclic_permutations.add(array2)

    return list(non_cyclic_permutations)

# Example usage:
length = 6
x = 3
result = generate_non_cyclic_permutations(length, x)
#print(f"All non-cyclic permutations of length {length} with {x} ones: {result}")

import cvxpy as cp
import numpy as np
d1 = cp.Variable([1,4], integer = True)
d2 = cp.Variable([1,4], integer = True)
A = cp.Variable(boolean = True)
B = cp.Variable(boolean = True)
C = cp.Variable(boolean = True)
D = cp.Variable(boolean = True)
choose = cp.Variable(4, boolean = True)


obj = sum(d1[0,:] + d2[0,:])


constraints = []


#for these constraints below A,B,C,D represent the permutation we will chose we make it so the di
array_1 = np.array(result[0]) # selectt the 4th array from our permutations
index_1 = np.where(array_1 == 1)[0] # gets the indexes of the values where busses should be 

constraints.append(d1[0,0] == choose[0]*index_1[1] - index_1[0])
constraints.append(d2[0,0] == choose[0]*index_1[2] - index_1[1])
    
    
array_2 = np.array(result[1])# selectt the 2nd array from our permutations
index_2 = np.where(array_2 == 1)[0] # gets the indexes of the values where busses should be 

constraints.append(d1[0,1] == choose[1]*index_2[1] - index_2[0])
constraints.append(d2[0,1] == choose[1]*index_2[2] - index_2[1])
    
array_3 = np.array(result[2])# selectt the 3rd array from our permutations
index_3 = np.where(array_3 == 1)[0] # gets the indexes of the values where busses should be 

constraints.append(d1[0,2] == choose[2]*index_3[1] - index_3[0])
constraints.append(d2[0,2] == choose[2]*index_3[2] - index_3[1])
    
array_4 = np.array(result[3]) # selectt the 1th array from our permutations
index_4 = np.where(array_4 == 1)[0] # gets the indexes of the values where busses should be 

constraints.append(d1[0,3] == choose[3]*index_4[1] - index_4[0])
constraints.append(d2[0,3] == choose[3]*index_4[2] - index_4[1])
    
constraints.append(sum(choose[:]) == 1)
   

problem = cp.Problem(cp.Maximize(obj), constraints)

problem.solve(solver = cp.GUROBI, verbose = False)

if(choose[0].value == 1):
    print("Best Permutation", result[0])

if(choose[1].value == 1):
    print("Best Permutation", result[1])

if(choose[2].value == 1):
    print("Best Permutation", result[2])

if(choose[3].value == 1):
    print("Best Permutation", result[3])


```
